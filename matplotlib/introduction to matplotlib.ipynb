### Which one should i use(pyplot or matplotlib(oo) method
 * when plotting something quickly it is okay to use the pyplot method
 * when plotting something more advances use oo method 

```python
heart_disease=pd.read_csv("6.1 heart-disease.csv")
heart_disease
```

```python
heart_disease.plot.hist()
```

```python
heart_disease.plot(figsize=(20,10),subplots=True);
```

```python
over_50= heart_disease[heart_disease["age"]> 50]
```

```python
over_50.head()
```

```python
# this is pyplot method
over_50.plot(kind="scatter",x="age",y="chol", c="target",figsize=(10,6));
```

```python
# this is madlibpyplot(OO) meythod object oriented 
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x ="age",y="chol", c="target",ax=ax);
ax.set_xlim(45,100)
```

```python
#Object Oriented from scratch 
fig, ax=plt.subplots(figsize=(10,7))
#plot the data
scatter= ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
# customize the plot
ax.set(title="Heart Disease and Cholestrol",
       xlabel="Age",
       ylabel="cholestrol")
                    
# add a legend
ax.legend(*scatter.legend_elements(), title="Target");
#@ add a horizontal line 
ax.axhline(over_50["chol"].mean(),
           linestyle='--')
```

```python
 #subplot of chol,age, thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                           sharex=True)
#add a data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                     c=over_50["target"])
                     
                    
#customize the data
ax0.set(title="heart disease and cholestrol",
     
        ylabel="chol")
# Adds a legend
ax0.legend(*scatter.legend_elements(), title="target")

# add a line 
ax0.axhline(over_50["chol"].mean(),
            linestyle='--')
#now same for ax1
ax1.scatter(x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"])
# customize the data
ax1.set(title="heart disease and thalach",
         xlabel="age",
       
        ylabel="thalach")
            
 #add a legend
ax1.legend(*scatter.legend_elements(), title="target")

#add a line
ax1.axhline(over_50["thalach"].mean(),
            linestyle='--')
#add a common title
fig.suptitle("Heart disease and Analysis",fontsize='20',fontweight='bold');
        
                    
```

# customization and other things

```python
 #subplot of chol,age, thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                           sharex=True)
#add a data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                     c=over_50["target"],
                   cmap="winter")
                     
                    
#customize the data
ax0.set(title="heart disease and cholestrol",
     
        ylabel="chol")
# removing the grid
ax0.set_xlim([50,80])
# Adds a legend
ax0.legend(*scatter.legend_elements(), title="target")

# add a line 
ax0.axhline(over_50["chol"].mean(),
            linestyle='--')
#now same for ax1
ax1.scatter(x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"],
           cmap="winter")
# customize the data
ax1.set(title="heart disease and thalach",
         xlabel="age",
       
        ylabel="thalach")
#removing the grid
ax1.set_xlim([50,80])
ax1.set_ylim([100,200])
            
 #add a legend
ax1.legend(*scatter.legend_elements(), title="target")

#add a line
ax1.axhline(over_50["thalach"].mean(),
            linestyle='--')
#add a common title
fig.suptitle("Heart disease and Analysis",fontsize='20',fontweight='bold');
        
                    
```
